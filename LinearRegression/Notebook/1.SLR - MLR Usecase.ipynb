#### Machine Learning

### Types of ML Problems

#### Supervised - Lables known 
- Regression - The Output variable is continuous.
    - Revenue prediction for sales
- Classification - The Output variable is categorical (discrete).
    - CI classification
    - Opportunity Prediction
- Time series forecasting
    - Backlog Forecasting

#### Unsupervised - Labels Unknown
- Clustering - Dividing the data points into a number of groups such that data points in the same groups are more similar to other data points in the same group and dissimilar to the data points in other groups.
- Anomaly Detection

#### Recommendation Engine
- Recommendation and Ranking

#### Optimization Problems
- Objective function to minimize or maximize subjected to some constraints.

```python
# importing required libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

#### Simple Linear Regression 

 y = mx + c , where y is the independent (predictors) variable and x is the dependent (target) variable.
 

```python
data_slr_0 = pd.read_csv('./Data/tvmarketing.csv')
data_slr_0.head()
```

```python
data_slr_0.columns = ['TV_adv_budget(L)','Sales(Cr)']
data_slr_0.head(2)
```

```python
# checking for missing data
data_slr_0.isna().any()
```

```python
data_slr_0.describe()
```

```python
data_slr_0.info()
```

```python
# assigning TV adv budget to x and Sales to y
x,y = data_slr_0['TV_adv_budget(L)'],data_slr_0['Sales(Cr)']
# Plotting a sccaterplot
plt.scatter(x,y)
plt.title("TV Adv vs Sales")
plt.xlabel("TV Adv Budget(Lacks)")
plt.ylabel("Sales in Cr")
```

```python
#getting the intecept and slope for the regression line
m, b = np.polyfit(x, y, 1)b
#plotting the regression line
plt.plot(x,m*x+b,color = 'r',label = 'Regression Line')
```

```python
#splitting the data into train and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 10)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
lin_mod = LinearRegression()
lin_mod.fit(x_train,y_train)
```

```python
#reshaping the arrays

x_train = x_train[:,np.newaxis]
y_train = y_train[:,np.newaxis]
# x_train = x_train.to_numpy().reshape(-1,1)
```

```python
x_train.shape,y_train.shape
```

```python
lin_mod.fit(x_train,y_train)
```

```python
lin_mod.coef_
```

```python
lin_mod.intercept_
```

```python
y_pred = lin_mod.predict(x_test[:,np.newaxis])
y_pred[:5]
```

```python
from sklearn.metrics import r2_score
print('R(squared) :',r2_score(y_test,y_pred))
```

```python
#plotting the Actual vs Predicted and the regression line for test data
c = [i for i in range(0,60)]
plt.scatter(c,y_test,color = 'grey',label = 'Actual')
plt.scatter(c,y_pred,color = 'green')
plt.title('Actual vs Predicted')

plt.plot(c,0.04474675*np.array(c)+7.30826074,color = 'r',label = 'Regression Line')
```

```python
data_sales_0 = pd.read_csv('./Data/advertising.csv') 
data_sales_0.head()
```

```python
sns.pairplot(data_sales_0)
```

```python
sns.heatmap(data_sales_0.corr(),annot= True)
```

```python
%matplotlib inline
plt.figure(figsize = [12,7])
plt.boxplot(data_sales_0,labels= ['TV','Radio',"Newspaper","Sales"])
```

```python
#calculating the upper and lower limit of values
Q1 = data_sales_0.Newspaper.quantile(0.25)
Q3 = data_sales_0.Newspaper.quantile(0.75)
IQR = Q3 -Q1
lower_limit = Q3 + 1.5*IQR
upper_limit = Q1 - 1.5*IQR
```

```python
#finding the outliers
data_sales_0[(data_sales_0.Newspaper <= upper_limit) | (data_sales_0.Newspaper >= lower_limit)]
```

```python
#filtering out the outliers
data_sales_1 = data_sales_0[~((data_sales_0.Newspaper <= upper_limit) | (data_sales_0.Newspaper >= lower_limit))]
data_sales_1.shape
```

```python
# Plotting the filtered data 
plt.figure(figsize = [10,8])
plt.boxplot(data_sales_1,labels= ['TV','Radio',"Newspaper","Sales"])
```

```python
#splitting the predictors and the target variable 
x = data_sales_1[["TV","Newspaper","Radio"]]
y = data_sales_1.Sales
```

```python
x.shape,y.shape
```

```python
#splitting the data into train and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state = 10)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
#fitting the linear model 
mlr_1 = LinearRegression()
mlr_1.fit(x_train,y_train.to_numpy().reshape(-1,1))
```

```python
mlr_1.coef_
```

```python
mlr_1.intercept_
```

```python
y_pred = mlr_1.predict(x_test)
```

```python
#getting the r2 value 
print('R2 Score :',r2_score(y_test,y_pred))
```

```python
#plotting actual vs predicted for test set 
plt.figure(figsize=[7,4])
c1 = [i for i in range(0,60)]
plt.plot(c1,y_test,color = 'green')
plt.plot(c1,y_pred,color = 'red')
plt.title("Actual vs Predicted")
plt.legend('AP')
```

```python
plt.figure(figsize=[6,4])
y_test.shape,y_pred.shape
plt.plot(c1,y_test.to_numpy().reshape(-1,1)-y_pred)
plt.title("Plot of Residuals")
plt.ylabel("Error")
```

```python
# Validating the model using statsmodels 
import statsmodels.api as sm
```

```python
# manually adding a constant as Statsmodels does not do it automatically 
x_train_sm = sm.add_constant(x_train)
x_train_sm.shape
```

```python
x_train_sm.head()
```

```python
sm_mlr_1 = sm.OLS(y_train,x_train_sm).fit()
sm_mlr_1.summary()
```

```python
# dropping the Newspaper Column
x_train_sm = x_train_sm.drop(["Newspaper"],axis = 1)
x_train_sm.head()
```

```python
sm_mlr_2 = sm.OLS(y_train,x_train_sm).fit()
sm_mlr_2.summary()
```

```python
#dropping Newspaper from x_test
x_test_sm = x_test.drop(["Newspaper"],axis=1)
x_test_sm = sm.add_constant(x_test_sm)
```

```python
#Plotting Actual vs Predicted 
plt.figure(figsize=[8,6])
plt.plot(c1,y_test,color = 'g')
plt.plot(c1,sm_mlr_2.predict(x_test_sm),color='grey')
plt.title("Actual vs Predicted")
plt.legend('AP')
plt.xlabel("Index")
plt.ylabel("Sales")
```

```python
#making a prediction for a new row
x_new = pd.Series([10,5,1])
x_new = x_new.to_numpy().reshape(-1,3)
sm_mlr_2.predict(x_new)
```
